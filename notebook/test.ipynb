```python
import pandas as pd

# 读取 Excel 文件
df = pd.read_excel("textx.xlsx")

# 接受用户输入
key = input("请输入关键词: ")

# 根据关键词查找行
filtered_df = df[df['Questions'].str.contains(key, case=False)]

# 打印对应的 Questions 和 Correct Answer 列的值
for index, row in filtered_df.iterrows():
    print("Question:", row['Questions'])
    print("Correct Answer:", row['Correct Answer'])
    print()

```
